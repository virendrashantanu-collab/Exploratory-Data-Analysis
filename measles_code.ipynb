```python
# Importing the modules required 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#Loading the measles dataset cases_year
data_year=pd.read_csv("cases_year.csv")
```

Creating Figure 1
Worldwide measles cases, rubella cases and total suspected measles and rubella cases by year

```python
#Summing measles cases by year to get global measles cases for each year and storing it as a dataframe
global_measles= data_year.groupby('year')['measles_total'].sum().to_frame()

#Summing rubella cases by year to get global rubella cases for each year and storing it as a dataframe
global_rubella= data_year.groupby('year')['rubella_total'].sum().to_frame()

# Summing measles+rubella cases by year to get global measles+rubella cases for each year and storing it as a dataframe
global_measles_rubella= data_year.groupby('year')['total_suspected_measles_rubella_cases'].sum().to_frame()

#Creating a line plot for worldwide measles, rubella and measles+rubella cases over the years

#Creating a list of years for x axis labels 
labels = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024', '2025']

#Setting figure size and selecting data required for the lineplot 
fig1, ax=plt.subplots(figsize=(10, 5))
ax.plot(labels, global_measles['measles_total'], label = "measles cases", linestyle="--")
ax.plot(labels, global_rubella['rubella_total'], label = "rubella cases", linestyle=":")
ax.plot(labels, global_measles_rubella['total_suspected_measles_rubella_cases'], label = "total suspected measles and rubella cases", linestyle="-")

#Setting the x axis name, y axis name and title of plot
ax.set_xlabel("Year")
ax.set_ylabel("Number of Cases")
ax.set_title('Worldwide measles and rubella cases')

# Removing the top boundary and right boundary of the figure
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)

#Adding legend
ax.legend()
plt.show()
```
