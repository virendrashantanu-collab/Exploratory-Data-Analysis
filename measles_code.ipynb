```python
# Importing the modules required 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

#Loading the measles dataset cases_year
data_year=pd.read_csv("cases_year.csv")
```

Creating Figure 1
Worldwide measles cases, rubella cases and total suspected measles and rubella cases by year

```python
#Summing measles cases by year to get global measles cases for each year and storing it as a dataframe
global_measles= data_year.groupby('year')['measles_total'].sum().to_frame()

#Summing rubella cases by year to get global rubella cases for each year and storing it as a dataframe
global_rubella= data_year.groupby('year')['rubella_total'].sum().to_frame()

# Summing measles+rubella cases by year to get global measles+rubella cases for each year and storing it as a dataframe
global_measles_rubella= data_year.groupby('year')['total_suspected_measles_rubella_cases'].sum().to_frame()

#Creating a line plot for worldwide measles, rubella and measles+rubella cases over the years

#Creating a list of years for x axis labels 
labels = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024', '2025']

#Setting figure size and selecting data required for the lineplot 
fig1, ax=plt.subplots(figsize=(10, 5))
ax.plot(labels, global_measles['measles_total'], label = "measles cases", linestyle="--")
ax.plot(labels, global_rubella['rubella_total'], label = "rubella cases", linestyle=":")
ax.plot(labels, global_measles_rubella['total_suspected_measles_rubella_cases'], label = "total suspected measles and rubella cases", linestyle="-")

#Setting the x axis name, y axis name and title of plot
ax.set_xlabel("Year")
ax.set_ylabel("Number of Cases")
ax.set_title('Worldwide measles and rubella cases')

# Removing the top boundary and right boundary of the figure
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)

#Adding legend
ax.legend()
plt.show()
```

Creating Figure 2 
Total suspected measles and rubella cases by region and year

```python
# Grouping the data by region and year to get total suspected measles and rubella cases for each region by year
regional_measles_rubella= data_year.pivot_table(index='year', columns= 'region', values='total_suspected_measles_rubella_cases', aggfunc='sum').reset_index()

# Creating a stacked bar plot to show the number of cases each year by region 
fig2, ax = plt.subplots()

ax.bar(x=regional_measles_rubella["year"], height=regional_measles_rubella["AFRO"], color="navy", label="Africa")
ax.bar(x=regional_measles_rubella["year"], height=regional_measles_rubella["SEARO"], bottom=regional_measles_rubella["AFRO"],
         color="skyblue", label="South-East Asia")
ax.bar(x=regional_measles_rubella["year"], height=regional_measles_rubella["WPRO"], bottom=regional_measles_rubella[["AFRO", "SEARO"]].sum(axis=1), 
        color="purple", label="Western Pacific")
ax.bar(x=regional_measles_rubella["year"], height=regional_measles_rubella["EMRO"], 
        bottom=regional_measles_rubella[["AFRO", "SEARO", "WPRO"]].sum(axis=1), 
        color="green", label="Eastern Mediterranean")
ax.bar(x=regional_measles_rubella["year"], height=regional_measles_rubella["EURO"], 
        bottom=regional_measles_rubella[["AFRO", "SEARO", "WPRO", "EMRO"]].sum(axis=1), 
        color="orange", label="Europe")
ax.bar(x=regional_measles_rubella["year"], height=regional_measles_rubella["AMRO"], 
        bottom=regional_measles_rubella[["AFRO", "SEARO", "WPRO", "EMRO", "EURO"]].sum(axis=1), 
        color="red", label="Americas")

#Setting the x axis name, y axis name, title of plot and ledgend
ax.set_xlabel("Year")
ax.set_ylabel("Number of Suspected Cases")
ax.set_title('Number of total suspected measles and rubella cases by region')
ax.legend(loc='best')

# Removing the top boundary and right boundary of the figure
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)
plt.show()
```

Creating Figure 3
Diagnoses method for measles by region

```python
# Grouping the data to get mean values of measles/rubella data by region
region_measles=data_year.groupby(['region']).mean(numeric_only=True).round().reset_index()

# Creating a stacked bar plot to identify the main method of diagnosis for measles in each region
fig3, ax = plt.subplots()
ax.bar(x=region_measles["region"], height=region_measles['measles_lab_confirmed'], color="red", label="Lab Confirmed")
ax.bar(x=region_measles["region"], height=region_measles['measles_epi_linked'], bottom=region_measles["measles_lab_confirmed"],
         color="green", label="EPI linked")
ax.bar(x=region_measles["region"], height=region_measles['measles_clinical'], bottom=region_measles[["measles_lab_confirmed", "measles_epi_linked"]].sum(axis=1), 
        color="orange", label="Clinically confirmed")

#Setting the x axis name, y axis, title and adding ledgend
ax.set_xlabel("Region")
ax.set_ylabel("Average measles cases")
ax.set_title('Diagnoses of measles cases by region')
ax.legend(loc='best')

# Removing the top boundary and right boundary of the figure
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)

# Adding a label on top of each bar indicating the avereage measles cases in the region
ax.bar_label(ax.containers[-1])
plt.show()
```

Figure 4 
Creating a choropleth map showing the countries with the highest average measles cases

```python
# Grouping data by iso3 country code to get mean measles/rubella data 
country_measles=data_year.groupby(['iso3']).mean(numeric_only=True).reset_index()

# Importing plotly.express
import plotly.express as px

# Creating a choropleth map to identify the countries with highest average measles cases 
map=px.choropleth(country_measles,locations="iso3",locationmode="ISO-3",color="measles_total", title="Average Measle Cases by Country (2012-2025)",
                  color_continuous_scale=px.colors.sequential.Reds)

# Renaming the ledgend to "Number of Cases"
map.update_layout(
    coloraxis_colorbar=dict(
        title="Number of Cases"
    )
)
map.show()
```

Figure 5 
Top 10 countries with the highest measles incidence rate in 2024

```python
# Grouping data to get yearly measles incidence rate by country
country_measles_incidence= data_year.pivot_table(index='country', columns='year', values='measles_incidence_rate_per_1000000_total_population')
# Extracting the top ten countries with the highest incidence rate in 2024 
top10_incidence_2024=country_measles_incidence[2024].nlargest(n=10)

# Plotting a horizontal bar plot of incidence rate by country
fig6= top10_incidence_2024.plot.barh()
fig6.set_xlabel("Measles incidence rate per 100,000")
fig6.set_ylabel("Country")
fig6.set_title("Top ten countries by measles incidence rate in 2024")
plt.show()
```

Figure 6
Creating a line graph depicting average measles cases by month for each region

```python
#Loading the dataset cases_month.css
data_month= pd.read_csv("cases_month.csv")
# Grouping data to get mean number of measles cases per month by region
monthly_measles_region=data_month.pivot_table(index='month', columns='region', values='measles_total', aggfunc='mean').reset_index()
```

```python
# Creating list for x axis labels
months=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
# Selecting data to plot a line graph of measles cases by month for each region
fig4, ax= plt.subplots(figsize=(10, 5))
ax.plot(months, monthly_measles_region["AFR"], label="Africa")
ax.plot(months, monthly_measles_region["AMR"], label="Americas")
ax.plot(months, monthly_measles_region["EMR"], label="Eastern Mediterranean")
ax.plot(months, monthly_measles_region["EUR"], label="Europe")
ax.plot(months, monthly_measles_region["SEAR"], label="South-East Asia")
ax.plot(months, monthly_measles_region["WPR"], label="Western Pacific")

#Setting x label, y label, titel and ledgend
ax.set_xlabel("Month")
ax.set_ylabel("Average measles cases")
ax.set_title("Monthly measles cases by region")
ax.legend()

#Removing top and right outline of graph
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)

plt.show()
```

Figure 7
Scatter plot of vaccination coverage against measles incidence 

```python
#Loading vaccination data
data_vaccination=pd.read_csv("measles_vaccination.csv")

# Renaming the regions such that the match the previous datasets 
data_vaccination_edit=data_vaccination.replace({'AFR': 'AFRO', 'EMR': 'EMRO', 'EUR': 'EURO','AMR': 'AMRO', 'SEAR': 'SEARO', 'WPR': 'WPRO'})
# Removing the last row as it is not relavent to the dataset
data_vaccination_edit = data_vaccination_edit[:-1]
# Converting the coloumn year to integer to remove the decimal 
data_vaccination_edit= data_vaccination_edit.astype({'YEAR':'int'})
```

```python
# Grouping the vaccination data to get vaccination coverage for each region by year
vaccination_grouped=data_vaccination_edit.pivot_table(index='CODE', columns= 'YEAR', values='COVERAGE').reset_index()
# Converting the grouped data into long format
vaccination_long=vaccination_grouped.melt(id_vars=["CODE"], var_name="YEAR", value_name="COVERAGE")
# Extracting the vaccination coverage coloumn
coverage= vaccination_long["COVERAGE"]

#Grouping the data_year dataset to get measles incidence for each region by year
measles_incidence_grouped=data_year.pivot_table(index='region', columns= 'year', values='measles_incidence_rate_per_1000000_total_population', aggfunc='mean').reset_index()
#Converting the grouped data into long format
measles_incidence_long=measles_incidence_grouped.melt(id_vars=["region"], var_name="year", value_name="measles_incidence_rate_per_1000000_total_population")
#Removing the last six rows corresponding to the year 2025 as the vaccination dataset contains data until 2024
measles_incidence_long= measles_incidence_long[:-6]
# Merging the coverage column extracted from the vaccination dataset to the incidence data
measles_incidence_long.insert(3, "Coverage", coverage)
```

```python
# Creating a scatter plot of vaccination coverage against measles incidence by year
scatter_plot= sns.relplot(data=measles_incidence_long, x='Coverage', y='measles_incidence_rate_per_1000000_total_population',
                          hue='year')
scatter_plot.set(xlabel='Measles 2nd dose vaccine coverage (%)', ylabel='Measles incidence rate per 1,000,000')
plt.show()
```
